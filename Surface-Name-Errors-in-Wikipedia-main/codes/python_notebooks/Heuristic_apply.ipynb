```python
"""
1. Load JSON of type [dest][SN] = count 
2. Apply your heuristic to divide it in 2 parts CorrectSN and IncorrectSN
3. Dump the following dicts:
    1. CorrectSN {dest : [SN list]}
    2. IncorrectSN {dest : [SN list]}
    3. CorrectSN_onlyForWhich_Incorrect_exist {dest : [SN list]}
"""
```

```python
file_path = "/ssd/Dumps/"
```

```python
import json

with open(file_path+'dest_SN_count.json','r') as file:
    D_act = json.load(file)
    file.close()

Correct_Surface_Names = {}
Wrong_Surface_Names = {}

for key,val in D_act.items():
    if(len(val)>3):
        Correct_Surface_Names[key] = {}
        Wrong_Surface_Names[key] = {}
        count = 0
        for k,v in val.items():
            count+=v
        # print(count)
        for k,v in val.items():
            try:
                if(v/count>=0.05):
                    Correct_Surface_Names[key][k] = (v,v/count) 
                else:
                    Wrong_Surface_Names[key][k] = (v,v/count)
            except:
                do_nothing = 1
                #print(key,i,k,v)
print("now dump")
with open(file_path+'Correct_Surface_Names.json','w', encoding='utf-8') as file:
    json.dump(Correct_Surface_Names, file,ensure_ascii=False)
    file.close()

with open(file_path+'Incorrect_Surface_Names.json','w', encoding='utf-8') as file:
    json.dump(Wrong_Surface_Names, file,ensure_ascii=False)
    file.close() 
```
